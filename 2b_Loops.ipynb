```python
#Prime Numbers
start = int(input("Enter the start of the range: "))
end = int(input("Enter the end of the range: "))

for num in range(start, end):    #steps through values in given range
    prime = True                #assumes it's prime
    for check in range(2,num):  #steps through each number smaller than num
        if num % check == 0:     #checks if it is divisible by anything
            prime = False       #if it is divisible, makes it not prime
    if prime:
      if num != 1:              #does not print if 1
        print(num, end=" ")              #displays prime number
```

```python
#Leap Year
while True:
  year = int(input("Please enter a year: "))
  if year % 4 != 0:
    YearType = "common"
  elif  year % 100 != 100:
    YearType = "leap"
  elif year % 400 != 0:
    YearType = "common"
  else:
    YearType = "leap"

  print(f"It is a {YearType} year.")
```

```python
#Lothar Collatz
while True:
  count=0
  c0 = int(input("Enter a non-zero, positive number: "))
  try:
    if c0 <= 0:
      print("Please enter a NON-ZERO, POSITIVE number!!! ")
    else:
      while c0 != 1:
        if c0 % 2 == 0:
          c0 = int(c0 / 2)
        else:
          c0 = 3 * c0 + 1
        print(c0)
        count+=1
      print(f"Number of Steps = {count}")
  except ValueError:
      print("Invalid input. Please enter a number.")
```

```python
#Break/Continue
#Even Numbers Only
for i in range(1,11):
  if i % 2 != 0:
    continue
  else:
    print(i)
```

```python
#FizzBuzz
for i in range(1,51):
  if i % 3 == 0 and i % 5 == 0:
    print("FizzBuzz")
  elif i % 3 == 0:
    print("Fizz")
  elif i % 5 == 0:
    print("Buzz")
  else:
    print(i)
```

```python
import random
targetNumber = random.randint(1, 100)
lives = 10

print(f"Welcome to my entirely random guessing game!\nYou have {lives} lives.")
guess = int(input("Guess a number between 1 and 100: "))

while guess!= targetNumber:
  if guess > targetNumber:
    guess = int(input(f"Too High!!!\nGuess again: "))
  elif guess < targetNumber:
    guess = int(input(f"Too Low!!!\nGuess again: "))
  lives-=1
  print(f"{lives} live(s) remaining.")
  if lives == 0:
    print("You ran out of lives!\nGAME OVER")
    break

if guess == targetNumber:
  print(f"Well done! You guessed it! It was {targetNumber}")
```

```python
#bmi
while True:
  try:
      weight = float(input("Enter your weight in kg: "))
      height = float(input("Enter your height in m: "))

      if height <= 0 or weight <= 0:
          print("Height and weight must be positive values.")
      else:
          bmi = weight / (height ** 2)

          if bmi < 18.5:
              category = "Underweight"
          elif 18.5 <= bmi < 25:
              category = "Normal weight"
          elif 25.0 <= bmi < 30:
              category = "Slight overhang"
          elif 30.0 <= bmi < 35:
              category = "Obese"
          else:
              category = "Clinically Obese"

          print(f"Summary:\nYour Height is: {height}\nYour Weight is: {weight}\nYour BMI is: {bmi:.2f}\nAssessment: {category}")
  except ValueError:
      print("Invalid input. Please enter numeric values for weight and height.")
```
