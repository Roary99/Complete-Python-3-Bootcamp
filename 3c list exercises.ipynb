```python
# Goldbachs conjecture
# Ross 04/12/25

# Create a display of prime pairs that illustrate Goldbachs conjecture :
start = int(input("Enter the start of the range: "))
end = int(input("Enter the end of the range: "))
evenNum= int(input("Enter an even number to test Goldbach's Conjecture: "))
primes =[]
for num in range(start, end + 1):
    prime = True
    if num == 1:
      continue
    for i in range(2,num):
        if num % i==0:
            prime = False
    if prime:
      primes.append(num)
print(f"The primes in the range {start}-{end} are: {', '.join(str(digit) for digit in primes)}")  #print list of primes in user defined range, joins list into a easy to read string
goldbach_pairs = []
for prime in primes: #loop through all primes
  if prime > evenNum // 2: #stop once prime is more than half the even number so no duplicates (x,y) and (y,x)
    break
  remainder = evenNum - prime
  if remainder in primes: #check if remainder after subtraction is a prime
    goldbach_pairs.append((prime, remainder))

print("Goldbach pairs:")
for pair in goldbach_pairs:  # Loop through all found pairs
    print(f"{evenNum} is the sum of {pair[0]} and {pair[1]}")
```
